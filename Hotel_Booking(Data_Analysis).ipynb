```python
## importing libraries##

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading the dataset.

df=pd.read_csv('hotel_booking.csv')
```

```python
# Exploratory Data analysis and Data cleaning .

df.head()
```

```python
df.info()
```

```python
df['reservation_status_date']=pd.to_datetime(df['reservation_status_date'])
```

```python
df.info()
```

```python
df=df.drop(['name','email','phone-number','credit_card'],axis=1)
```

```python
df.shape
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(['company','agent'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.describe(include='object')
```

```python
for col in df.describe(include='object').columns:
  print(col)
  print(df[col].unique())
  print('-'*50)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['adr'].plot(kind='box')
```

```python
df=df[df['adr']<400]
```

```python
# Data Analysis and Visulization

cancelled_percentage=df['is_canceled'].value_counts(normalize=True)
print(cancelled_percentage)
```

```python
plt.figure(figsize=(3,4))
plt.title('Reservation status count')
plt.bar(['not cancelled','cancelled'],df['is_canceled'].value_counts(),edgecolor='k',width=0.7)
plt.show()
```

```python
plt.figure(figsize=(3,4))
x=sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
legend_labels,_=x.get_legend_handles_labels()
#x.legend(bbox_to_anchor(1,1))
plt.title('Reservationstatus in differnt hotel',size=10)
plt.xlabel('hotel')
plt.ylabel('number of reservation')
plt.show()
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(5,4))
x=sns.countplot(x='month',hue='is_canceled',data=df,palette='bright')
legend_labels,_=x.get_legend_handles_labels()
#x.legend(bbox_to_anchor(1,1))
plt.title('Reservation ststus for each month',size=15)
plt.xlabel('month')
plt.ylabel('number of resrvation')
plt.legend(['not canceled','canceled'])
plt.show()
```
